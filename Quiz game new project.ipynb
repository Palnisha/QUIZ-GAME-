# QUIZ GAME BY USING OOPS CONCEPT 


 QUIZ GAME BY USING DIFFERNTE GK  QUESTIONS

```python
###by using oops concept:

###### using define function calling the options and answers

class Question:
    def __init__(self, prompt, options,answer):
        self.prompt = prompt
        self.options =options
        self.answer = answer

## giving index  for the user to choose their answers:
        class Game:
    def __init__(self):
        self.questions = []
        self.current_question_index = 0
        self.score = 0

## adding the question to the quize game:

    def add_question(self, prompt, options,answer):
        question = Question(prompt,options, answer)
        self.questions.append(question)
    
#### display the question by using enumerate function:

    def display_question(self):
        question = self.questions[self.current_question_index]
        print(question.prompt)
        for i, option in enumerate(question.options):
            print(f"{i + 1}. {option}")
            
 ## by user define function getting  user answers:

    def get_user_answer(self):
        while True:
            try:
                user_choice = int(input("Enter your choice (1-4): "))
                if user_choice < 1 or user_choice > 4:
                    raise ValueError()
                return user_choice
            except ValueError:
                print("Invalid input. Please enter a number between 1 and 4.")
                
### checking answers from user answers:

    def check_answer(self, user_answer):
        question = self.questions[self.current_question_index]
        if user_answer == question.answer:
            self.score += 1
            
 ##### display next question after user answer:

    def next_question(self):
        self.current_question_index += 1
        
#### Run the game:

    def run_game(self):
        print("Welcome to Quiz Game.\n")
        
        while self.current_question_index < len(self.questions):
            self.display_question()
            user_answer = self.get_user_answer()
            self.check_answer(user_answer)
            self.next_question()
        
        self.display_score()
        
        
 ### display the score for the user :

    def display_score(self):
        print("\nQuiz completed!")
        print("Your score:", self.score, "/", len(self.questions))
    
    
    

# Create a quiz object:

game = Game()

# Add questions to the quiz game:

game.add_question('What is the capital of France? ',['A. amerika','B. paris', 'C. londan'], 2),'\n'
game.add_question('What is the largest planet in our solar system? ',['A. Mars', 'B. Neptune', 'C. jupiter'] ,3),'\n'
game.add_question('Who painted the Mona Lisa?',['A. Einsten', 'B. Leonard da vinci', 'C. cazon'],2)
game.add_question('The Centre for Cellular and Molecular Biology is situated at :',['A.Patna','B.Jaipur','C.Hyderabad'],3)
game.add_question('Where is the Railway Staff College located? :',['A.Delhi','B.vadodra','C.Hyderabad'],2)
game.add_question('famous Dilwara Temples are situated in:',['A.Rajsthan','B.UP','C.Maharashtra'],1)
game.add_question('Wadia Institute of Himalayan Geology is located at :',['A.Delhi','B.Shimla','C.Dehradun'],3)
game.add_question('Bijapure is known for its :',['A.severe drought condition','B.Gol Gumbaz','C.Heavy rainfall'],2)
game.add_question('The members of the Rajya Sabha are elected by:',['A.people','B.Loksaba','C.elected member of the legislative assembly'],3)
game.add_question('The power to decide an election petition is vested in the:',['A.High courts','B.supreme courts','C.parliament'],1)


# Run the game
game.run_game()
```
